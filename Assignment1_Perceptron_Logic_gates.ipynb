```python
import random

def step_function(x):
    if x >= 0:
        return 1
    else:
        return 0

class Perceptron:

    def __init__(self, lr=0.1, epochs=20):
        self.lr = lr
        self.epochs = epochs
        self.weights = None
        self.bias = None

    def train(self, X, y):
        n_samples = len(X)
        n_features = len(X[0])

        self.weights = [random.uniform(-1, 1) for _ in range(n_features)]
        self.bias = random.uniform(-1, 1)

        for _ in range(self.epochs):
            for i in range(n_samples):
                linear_output = sum(X[i][j] * self.weights[j] for j in range(n_features)) + self.bias
                y_pred = step_function(linear_output)

                error = y[i] - y_pred

                for j in range(n_features):
                    self.weights[j] = self.weights[j] + self.lr * error * X[i][j]
                self.bias = self.bias + self.lr * error

    def predict(self, X):
        predictions = []
        for x in X:
            linear_output = sum(x[j] * self.weights[j] for j in range(len(x))) + self.bias
            predictions.append(step_function(linear_output))
        return predictions


X = [
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]

AND_y = [0,0,0,1]
OR_y = [0,1,1,1]
NAND_y = [1,1,1,0]
NOR_y = [1,0,0,0]


gates = {
    "AND": AND_y,
    "OR": OR_y,
    "NAND": NAND_y,
    "NOR": NOR_y
}


for gate in gates:

    print("\n",gate,"Gate")

    model = Perceptron(lr=0.1, epochs=25)
    model.train(X, gates[gate])

    predictions = model.predict(X)

    for i in range(len(X)):
        print(X[i][0], X[i][1], "->", predictions[i])
```
